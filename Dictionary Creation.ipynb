```python
# Importing the original table
import csv
import pandas as pd

panama_list = pd.read_csv("Merged/Original Panama Names for the Dictionary.csv", encoding='ISO-8859-1')
panama_list = panama_list.values.tolist()
#print(panama_list)
```

```python
# Importing thee list of unique names

unique = pd.read_excel("Merged/Unique list of names.xlsx")
unique = unique.values.tolist()
#print(unique)
```

```python
# Finding names akin to Bloomberg names

for q in range(2254, len(unique)):
    repeat = 1
    while repeat != 0:
        print(q, "Panama name:    ", unique[q])
        bloomberg = input("Enter a string of words: ")
        bloomberg_list = bloomberg.split()
        good = []
        for i in range(len(panama_list)):
            in_panama = True
            panama_low = str(panama_list[i][0]).lower()
            for j in range(len(bloomberg_list)):
                if bloomberg_list[j] not in panama_low:
                    in_panama = False
            if in_panama == True:
                separate_name = True
                for j in range(len(bloomberg_list)):
                    index = panama_low.index(bloomberg_list[j])
                    alphabet = [chr(p) for p in range(ord('a'),ord('z')+1)]
                    if ((index != 0) and (panama_low[index-1] in alphabet)) or ((index + len(bloomberg_list[j]) != len(panama_low) and (panama_low[index+len(bloomberg_list[j])] in alphabet))):
                        separate_name = False
                if separate_name == True:
                    good.append([panama_list[i][0], panama_list[i][1], panama_list[i][2]])
        good_table = pd.DataFrame(good)
        print(good_table)
        repeat = int(input("Once again?   "))
    hurry_up = input("Hurry up!   ")
    
    # Importing a temporary file

    data = pd.read_excel("Merged/Temporary file for dictionary creation.xlsx", header = None)
    data = data.values.tolist()
    data_new = [[" ", " ", " ", " ", " "]]
    for i in range(len(data)):
        not_nan = False
        row = [unique[q][0], unique[q][1], " ", " ", " "]
        for j in range(len(data[i])-1,0, -1):
            if (str(data[i][j]) != "nan") and (not_nan == False):
                not_nan = True
                row[2], row[3], row[4] = str(data[i][1]), data[i][j-1], data[i][j]
                for k in range(2, j-1):
                    row[2] = row[2] + " " + str(data[i][k])
        data_new.append(row)     
    print("  ")

    with open('Merged/Perfect dictionary.csv', 'a') as f:
        writer = csv.writer(f)
        for i in range(len(data_new)):
            writer.writerow(data_new[i])            
```
