```python
# Importing a .csv file with Panama names and countries

import pandas as pd
import numpy as np

table = pd.read_csv('panama_officers_countries.csv', encoding = "ISO-8859-1")
table_list = table.values.tolist()
print(len(table_list))
```

```python
#  Importing tables with Datastream-Panama merged names

table_full = pd.read_csv('Merged/full_match.csv', encoding = "ISO-8859-1")
table_full = table_full.values.tolist()
table_almost_full = pd.read_csv('Merged/almost_full_match.csv', encoding = "ISO-8859-1")
table_almost_full = table_almost_full.values.tolist()
table_part = pd.read_csv('Merged/part_match.csv', encoding = "ISO-8859-1")
table_part = table_part.values.tolist()
print(len(table_full), len(table_almost_full), len(table_part))
#print(table_full)
```

```python
# Creating a set of merged names

panama_names = set()
for i in range(len(table_full)):
    panama_names.update({table_full[i][0]})
print(len(panama_names))
for i in range(len(table_almost_full)):
    panama_names.update({table_almost_full[i][0]})
print(len(panama_names))     
for i in range(len(table_part)):
    panama_names.update({table_part[i][0]})    
print(len(panama_names)) 
```

```python
# Creating a table with names not merged and countries not China

table1 = []
for i in range(len(table_list)):
    if (table_list[i][0] not in panama_names) and (table_list[i][1] != "China"):
        table1.append(table_list[i])
print(len(table1))        
```

```python
# Exporting the table with names not merged and countries not China

import csv

with open("panama_officers_countries_no_merged.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerows(table1)
```

```python
# Importing a list with no "mr." etc. after deleting them in Excel

table_new = pd.read_csv('panama_officers_countries_no_merged_no_mr.csv', encoding = "ISO-8859-1")
table_new = table_new.values.tolist()
print(len(table_new))
print(table_new)
```

```python
# Creating a table without repeated names

# DO NOT EXECUTE BY CHANCE!

table_super_new = []
for i in range(len(table_new)):
    a = True
    for j in range(len(table_super_new)):
        if table_new[i][0] == table_super_new[j][0]:
            a = False
    if a == True:
        table_super_new.append(table_new[i])
print(len(table_super_new))        
```

```python
# Creating a dictionary with countries as keys and names as values

dict = {}
for i in range(len(table_super_new)):
    if table_super_new[i][1] in dict:
        dict[table_super_new[i][1]].append([table_super_new[i][0], ""])
    else:
        dict[table_super_new[i][1]] = [[table_super_new[i][0], ""]]
print(dict)        
```

```python
# Creating separate .csv files for each country

import csv

not_identified = []
for i in dict:
    if (str(i) != 'nan'):
        title = "Panama by Country/CSV/" + i + ".csv"
        with open(title, "w") as f:
            writer = csv.DictWriter(f, fieldnames = ["Name", "Void"])
            writer.writeheader()
            writer = csv.writer(f)
            writer.writerows(dict[i])
#with open("Panama by Country/CSV/Iceland.csv", "w") as f:
#    writer = csv.writer(f)
#    writer.writerows(dict["Iceland"]) 
#print(dict["Iceland"])    
```

```python
# Creating a table with numbers of names for each country

count = []
for i in dict:
    count.append([i, len(dict[i])])
print(count)    
```

```python
# Exporting the table with counted countries

with open("Panama by Country/CSV/Summary.csv", "w") as f:
    writer = csv.DictWriter(f, fieldnames = ["Country", "Number"])
    writer.writeheader()
    writer = csv.writer(f)
    writer.writerows(count)
```
